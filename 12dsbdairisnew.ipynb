```python
fig,axes=plt.subplots(2,2,figsize=(16,8))

axes[0,0].set_title("Distribution of first column")
axes[0,0].hist(df["sepal_length"])

axes[0,1].set_title("Distribution of second column")
axes[0,1].hist(df["sepal_width"])

axes[1,0].set_title("Distribution of third column")
axes[1,0].hist(df["petal_length"])


axes[1,1].set_title("Distribution of fourth column")
axes[1,1].hist(df["petal_width"])
```

```python


# Boxplot for Sepal Length
sns.boxplot(y=df['sepal_length'])
plt.title("Boxplot - Sepal Length")
plt.show()

# Boxplot for Sepal Width
sns.boxplot(y=df['sepal_width'])
plt.title("Boxplot - Sepal Width")
plt.show()

# Boxplot for Petal Length
sns.boxplot(y=df['petal_length'])
plt.title("Boxplot - Petal Length")
plt.show()

# Boxplot for Petal Width
sns.boxplot(y=df['petal_width'])
plt.title("Boxplot - Petal Width")
plt.show()
```

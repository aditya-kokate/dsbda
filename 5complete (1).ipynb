```python
species_map = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
df['species_numeric'] = df['species'].map(species_map)
```

```python
species_numeric_list = df['species_numeric'].tolist()
```

```python
print("Numeric list for species column:")
print(species_numeric_list[:150])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["species"]=le.fit_transform(df["species"])
```

```python
grp=pd.DataFrame(df.groupby("species").agg(['mean','median', 'min', 'max', 'std']))
```
